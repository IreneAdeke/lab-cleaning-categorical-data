```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
customer_df = pd.read_csv("https://raw.githubusercontent.com/ironhack-labs/lab-cleaning-categorical-data/master/files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv")
```

```python
customer_df
```

```python
#Renaming the columns so they follow the PE8 (snake case)
#Putting an underscore where theres a space
customer_df.columns = customer_df.columns.str.replace(' ', '_')
 
```

```python
customer_df
```

```python
#We can drop the column Customer 
customer_df1 = customer_df.drop(columns = 'Customer')
```

```python
customer_df
```

```python
numerical_df = customer_df.select_dtypes(np.number)
```

```python
categorical_df = customer_df.select_dtypes(object)
categorical_df
```

```python
for i in categorical_df: 
    sns.catplot(x=i, y="Total_Claim_Amount", kind="box", data=customer_df)
    plt.show()
```

1. The Total Claim Amount is almost the same for customers with Medium Size, Small and Large Vehicle Size.
2. Customers with Luxury cars and Luxury SUV has a higher Total Claim Amount. The lowest own a Two and Four Door car.
3. The Total Claim Amount is the same for all Sales Channels. 
4. Customers with Tenew Offer Type 1 have a higher TCA(Total Claim Amount)
5. Single customers have highest TCA
6. THose with lower TCA live in rural areas while those with higher TCA live in Sub-urban areas. 
7. The difference between TCA for males and females is slight. 
8. Unemployed customers have higher TCA
9. Customers with Premium Coverage have a higher TCA than those with Basic Coverage
10. TCA for Yes and No responses is almost the same.
